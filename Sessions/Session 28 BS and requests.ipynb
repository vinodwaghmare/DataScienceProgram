```python
import requests
form bs4 import BeautifulSoup
```

```python
def get_data(soup):
    cards = soup2.find_all("div",attrs = {"class":"_1UoZlX"})
    for card in cards:
        title = card.find("div", attrs = {"class":"_3wU53n"})
        price = card.find("div",attrs = {"class":"_2rQ-NK"})
        rating = card.find("div",attrs = {"class":"hGSR34"})
        if rating and title and price:
            rating_text = rating.text
            title_text = title.text
            price_text = price.text
        else:
            rating_text = title_text= price_text =None
            
        print("{} {} {}".format(title.text,price.text,rating.text))

start_url = "https://www.flipkart.com/search?q=mobile&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&as-pos=0&as-type=HISTORY"
HOME_URL = "https://www.flipkart.com"

count = 0
while count < 10:
    print("**************{}*************".format(count))
    response = requests.get(start_url)
    soup = BeautifulSoup(response.content,"html.parser")
    get_data(soup)
    next_link = soup.find("a",text = "Next")

    start_url = HOME_URL + next_link.get("href")
    count += 1
```
