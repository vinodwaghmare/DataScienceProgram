```python
import json

fp = open("gloss.json","r")
content = fp.read()
d = json.loads(content)
# print(d)
d["glossary"]["title"] = "Have a Good Day"
d["glossary"]["GlossDiv"]["GlossList"]["GlossEntry"]["ID"] = ("HTML","SGML")
# print(d)
fp.close()

j = json.dumps(d,indent = 4)
fp = open("gloss.json","w")
fp.write(j)
fp.close()
```

```python
import xmltodict

fp = open("gloss.xml","r")
content = fp.read()
# print(content)
d = xmltodict.parse(content)
# print(d)
fp.close()

d["glossary"]["title"] = "Have a Good Day"
d["glossary"]["GlossDiv"]["GlossList"]["GlossEntry"]["@ID"] = ("HTML","SGML")
# print(d)

fp = open("gloss.xml","w")
x = xmltodict.unparse(d,indent = 4)
fp.write(x)
fp.close()
```
