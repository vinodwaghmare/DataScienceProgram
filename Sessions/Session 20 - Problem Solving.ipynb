Put line breaks in the given sequence such that the lines are printed nearly.

Assume that the length of each word is smaller than the line width.

```python
remaining_str = ""
output = []
fp = open("blog1.txt","r")
for line in fp:
    next_line =  remaining_str + line
    if  len(next_line) > 100:
        output.append(next_line[:100])
        remaining_str = next_line[100:]
         
    else: 
        output.append(next_line)
        remaining_str = ""

print(output)
        
    
for val in output:
    print(len(val))   
    
    
fp.close()

fp2 = open("blog8.txt","w")
fp2.writelines(output)
fp2.close()    
    
    
```
